```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
articles = requests.get(query_url).json()
article_list_10 = articles["response"]["docs"][0:500]

for article in article_list_10:
    # Run Vader Analysis on each Sample
    a = article['snippet']
    results = analyzer.polarity_scores(a)
    compound = results["compound"]
    pos = results["pos"]
    neu = results["neu"]
    neg = results["neg"]

    # Print Samples and Analysis
    print(a)
    print("Compound Score:", compound)
    print("Positive Score:", pos)
    print("Neutral Score:", neu)
    print("Negative Score: ", neg)
    print()
```
