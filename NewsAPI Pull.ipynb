```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
articles = requests.get(query_url).json()

article_list_10 = articles["articles"][0:16]

for article in article_list_10:
    a1 = article['content']
    # Run Vader Analysis on each Sample
    results = analyzer.polarity_scores(a1)
    compound = results["compound"]
    pos = results["pos"]
    neu = results["neu"]
    neg = results["neg"]
    
    
    # Print Samples and Analysis
    #print(article["title"])
    print("Compound Score:", compound)
    print("Positive Score:", pos)
    print("Neutral Score:", neu)
    print("Negative Score: ", neg)
    
    print()

    
```

```python
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
for article in article_list_10:
    a1 = article['content']
    # Run Vader Analysis on each Sample
    results = analyzer.polarity_scores(a1)
    compound = results["compound"]
    pos = results["pos"]
    neu = results["neu"]
    neg = results["neg"]
    
    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)
    
    # Print Samples and Analysis
    #print(article["title"])
print("Mean Sentiment Analysis for News API users")
print("-------------------------------------------")
print("Compound Score:", np.mean(compound_list))
print("Positive Score:", np.mean(positive_list))
print("Neutral Score:", np.mean(neutral_list))
print("Negative Score: ", np.mean(negative_list))
print()
    
```
